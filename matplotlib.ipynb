```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,5,7]
y=[23,45,89]

x2=[3,5,7]
y2=[8,5,9]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,5,7]
y=[8,10,7]

x2=[3,5,7]
y2=[8,5,9]

plt.plot(x,y,label='first line')
plt.plot(x2,y2,label='second line')
plt.ylabel('This is Y label')
plt.xlabel('This is X label')
plt.title('Sample Chart')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x=[1,5,7]
y=[8,10,7]

x2=[3,5,7]
y2=[8,5,9]

plt.plot(x,y,label='first line',linewidth=5)
plt.plot(x2,y2,label='second line',linewidth=4)
plt.ylabel('This is Y label')
plt.xlabel('This is X label')
plt.title('Sample Chart')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x=[1,5,7]
y=[8,10,7]

x2=[3,5,7]
y2=[8,5,9]

plt.scatter(x,y,label='first line',linewidth=5)
plt.scatter(x2,y2,label='second line',linewidth=4)
plt.ylabel('This is Y label')
plt.xlabel('This is X label')
plt.title('Sample Chart')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x=[1,5,7]
y=[8,10,7]

x2=[3,5,7]
y2=[8,5,9]

plt.bar(x,y,label='first line',linewidth=5)
plt.bar(x2,y2,label='second line',linewidth=4)
plt.ylabel('This is Y label')
plt.xlabel('This is X label')
plt.title('Sample Chart')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]

x2=[3,7,3]
y2=[5,8,9]

plt.hist(x,y,label='first line',linewidth=5)
plt.hist(x2,y2,label='second line',linewidth=4)

plt.title('Sample Chart')
plt.legend()
plt.show()
```

```python
from cycler import cycler

import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['lines.linewidth']=2
plt.rcParams['lines.linestyle']='--'
data=np.random.randn(50)
plt.plot(data)
```

```python
plt.rcParams['axes.prop_cycle']=cycler(color=['r','g','b','y'])
plt.plot(data)   # first color is red
```
